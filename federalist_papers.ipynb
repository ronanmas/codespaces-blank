```python
PUNCS = ["!","$","-","?","\'",",",".",":",";","\""]

def cleaner(some_list,clean_list):
    for item in some_list:
        if item:
            if len(item) > 1:
                if item[-2]  in PUNCS:
                    item = item[:-2] + item[-1]
            if item[0] in PUNCS:
                item = item[1:]
            if item:
                if item[-1] in PUNCS:
                    item = item[:-1]
            lower_items = item.lower()
            clean_list.append(lower_items)

    return clean_list


def counter(clean_list, some_dict):
    for word in clean_list:
        if word in some_dict:
            some_dict[word] += 1
        else:
            some_dict[word] = 1
    return some_dict





def chi(expect_author,real_author, expect_dis, real_dis):
    chi = (((expect_author - real_author)**2)/ expect_author) + (((expect_dis - real_dis)**2)/ expect_dis)
    return chi



        
```
